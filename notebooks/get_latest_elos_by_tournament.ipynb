# Final dataset

```python
import pandas as pd
```

```python
years = [2020,2021,2022,2023]
for year in years:
    df = pd.read_csv(f'elos_{year}.csv',sep=';')
    df.drop_duplicates(subset=['leagueLabel','stageName','stageTournament','year','team'],keep='last',inplace=True)
    df.sort_values(by=['leagueLabel','stageTournament','stageName','rating'],inplace=True,ascending=False)
    df.to_csv(f'full_elo_ratings_{year}.csv',sep=';',index=False)
```
