# Preparing Oracle Elixir's data for final processing

```python
import pandas as pd
import numpy as np
```

```python
oe_df = pd.read_csv('oracles_elixir_lpl_data.csv',sep=';')
oe_df.columns
```

```python
oe_df['year'] = oe_df.gameDate.apply(lambda x: x[:4])
```

```python
oe_df.drop(columns=['GoldDiff15Top', 'GoldDiff15Jg', 'GoldDiff15Mid', 'GoldDiff15AD',
       'GoldDiff15Sup', 'BlueAssists15', 'BlueDeaths15', 'RedAssists15',
       'RedDeaths15'],inplace=True)

oe_df['playoffs'] = oe_df['playoffs'].apply(lambda x: 'Playoffs' if x == 1 else 'Regular Season')
oe_df['playoffs'].unique()
```

```python
null_tourneys = oe_df.split.isnull()
oe_df.loc[oe_df.split.isnull(),'playoffs']='Playoffs'
oe_df.loc[oe_df.split.isnull(),'split']='Regional Qualifier'
oe_df[null_tourneys]
```

```python
game_identifier = oe_df[['esportsPlatformId','gameNumber','split','playoffs','year','BlueTeamName','BlueTeamID','RedTeamName','RedTeamID']].copy()
game_identifier['leagueLabel'] = 'LPL'

lpl_code = {
    "FunPlus Phoenix": "FPX",
    "Team WE": "WE",
    "Anyone's Legend": "AL",
    "Bilibili Gaming": "BLG",
    "JD Gaming": "JDG",
    "LGD Gaming": "LGD",
    "Rare Atom": "RA",
    "Ultra Prime": "UP",
    "EDward Gaming": "EDG",
    "Top Esports": "TES",
    "ThunderTalk Gaming": "TTG",
    "LNG Esports(LI-NING GAMING)": "LNG",
    "LNG Esports": "LNG",
    "Invictus Gaming": "IG",
    "Weibo Gaming": "WBG",
    "Ninjas in Pyjamas": "NIP",
    "Oh My God": "OMG",
    "Royal Never Give Up": "RNG",
    "Suning":"SNG",
    "eStar":"ES",
    "Rogue Warriors":"RW",
    "Vici Gaming":"VG",
    "Dominus Esports":"DMO",
    "Victory Five":"V5"
}

game_identifier['team_code_blue'] = game_identifier['BlueTeamName'].apply(lambda x: lpl_code[x])
game_identifier['team_code_red'] = game_identifier['RedTeamName'].apply(lambda x: lpl_code[x])
```

```python
oe_df['BlueDeathsADR'] = oe_df['BlueDeathsEnd'].replace(0.0,1.0)
oe_df['RedDeathsADR'] = oe_df['RedDeathsEnd'].replace(0.0,1.0)
```

```python

oe_features = pd.DataFrame()
oe_features['winner'] = oe_df['winner']
oe_features['GoldDiffPerMinEnd'] = (oe_df['BlueTotalGoldEnd']-oe_df['RedTotalGoldEnd'])/(oe_df['gameDuration']/60)
oe_features['ADRDiffEnd'] = ((oe_df['BlueAssistsEnd']/oe_df['BlueDeathsADR'])-(oe_df['RedAssistsEnd']/oe_df['RedDeathsADR']))
oe_features['VisionScoreDiffPerMin'] = (oe_df['VisionScoreBlue']-oe_df['VisionScoreRed'])/(oe_df['gameDuration']/60)
oe_features['ObjectiveDiff'] = oe_df['BlueTowerKillsEnd']+oe_df['BlueInhibKillsEnd']+oe_df['BlueBaronKillsEnd']\
+oe_df['BlueDragonKillsEnd']-oe_df['RedTowerKillsEnd']-oe_df['RedInhibKillsEnd']-oe_df['RedBaronKillsEnd']\
-oe_df['RedDragonKillsEnd']
oe_features.fillna(axis=1,value=0,inplace=True)
oe_features['ObjectiveDiff'] = oe_features['ObjectiveDiff'].astype(int)
oe_features.head()
```

```python
oe_ready_df = pd.concat([game_identifier,oe_features],axis=1)
oe_ready_df['BlueTeamName'] = oe_ready_df['BlueTeamName'].replace('LNG Esports(LI-NING GAMING)','LNG Esports')
oe_ready_df['RedTeamName'] = oe_ready_df['RedTeamName'].replace('LNG Esports(LI-NING GAMING)','LNG Esports')
oe_ready_df['gameDate']=oe_df['gameDate']
oe_ready_df.to_csv('full_lpl_data_process.csv',sep=';',index=False)
```
