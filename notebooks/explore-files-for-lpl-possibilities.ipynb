# Checking for potential LPL files in our mappings

```python
import json
```

```python
list_of_games = []

with open('game_file_names_in_cloud.txt','r') as file_game_list:
    for game_line in file_game_list:    
        list_of_games.append(game_line[30:].lstrip().replace('.json.gz\n',''))
```

```python
list_of_games
```

```python
# Mapped games:
mapped_game_list = []

with open('esports-data/mapping_data.json','r') as mapping_file:
    mappings = json.load(mapping_file)
    for mapped_game in mappings:
        game_id = mapped_game.get('platformGameId',None)
        if game_id is not None:
            mapped_game_list.append(game_id)
```

```python
mapped_game_list
```

```python
difference_gamelist = list(set(list_of_games) - set(mapped_game_list))
len(difference_gamelist)
```

```python
difference_gamelist
```

```python
difference_maplist = list(set(mapped_game_list) - set(list_of_games))
print(len(difference_maplist))
difference_maplist
```

For all intents and purposes, that amount of files is not enough for an LPL season, let alone more. In this notebook, I conclude that I will need external data to cover the LPL's gaps.
